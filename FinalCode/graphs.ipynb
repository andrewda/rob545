```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

baseline = 3997

data_mapdist = pd.read_csv('mapdistPerGen.csv')
data_survivors = pd.read_csv('survivors.csv')

plt.figure(figsize=(8, 6))

# Plot baseline
plt.plot(data_mapdist.columns.astype(int), [baseline] * len(data_mapdist.columns), label='Baseline (Mean)', color='black', linestyle='dashed', alpha=0.5)

# Plot for mapdistPerGen.csv
# plt.figure(figsize=(7, 7))
for i in range(data_mapdist.shape[0]):
    plt.plot(data_mapdist.columns.astype(int), data_mapdist.iloc[i], label=f'Track {i+1}')

plt.title('Map Distance per Generation')
plt.xlabel('Generation')
plt.ylabel('Distance')
plt.legend(title='Track')
plt.grid(True)
plt.show()

# Plot for survivors.csv
# plt.figure(figsize=(7, 7))
for i in range(data_survivors.shape[0]):
    plt.plot(data_survivors.columns.astype(int), data_survivors.iloc[i], label=f'Track {i+1}')

plt.title('Survivors per Generation')
plt.xlabel('Generation')
plt.ylabel('Number of Survivors')
plt.legend(title='Track')
plt.grid(True)
plt.show()
```
